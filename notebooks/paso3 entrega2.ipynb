```python
!pip install great-expectations==0.18.21 -q

import pandas as pd
import great_expectations as gx

# =====================================================
# 1. CARGAR DATASET FINAL
# =====================================================
df = pd.read_csv("04_dataset_final_enriquecido.csv", dtype={"cod_dpto": str})

print("Dimensiones del dataset:", df.shape)
display(df.head())
print("\nNulos por columna:")
print(df.isnull().sum())

# =====================================================
# 2. CREAR VALIDATOR
# =====================================================
validator = gx.from_pandas(df)

# =====================================================
# 3. EXPECTATIONS
# =====================================================
resultados = {}

resultados["cod_dpto_no_nulo"] = validator.expect_column_values_to_not_be_null("cod_dpto")
resultados["dpto_no_nulo"] = validator.expect_column_values_to_not_be_null("dpto")
resultados["departamento_limpio_no_nulo"] = validator.expect_column_values_to_not_be_null("departamento_limpio")
resultados["anio_no_nulo"] = validator.expect_column_values_to_not_be_null("anio")
resultados["total_casos_no_nulo"] = validator.expect_column_values_to_not_be_null("total_casos")

resultados["anio_rango"] = validator.expect_column_values_to_be_between(
    "anio",
    min_value=2010,
    max_value=2025
)

resultados["total_casos_positivo"] = validator.expect_column_values_to_be_between(
    "total_casos",
    min_value=0
)

resultados["casos_femenino_positivo"] = validator.expect_column_values_to_be_between(
    "casos_femenino",
    min_value=0
)

resultados["casos_masculino_positivo"] = validator.expect_column_values_to_be_between(
    "casos_masculino",
    min_value=0
)

resultados["casos_adultos_positivo"] = validator.expect_column_values_to_be_between(
    "casos_adultos",
    min_value=0
)

resultados["casos_menores_positivo"] = validator.expect_column_values_to_be_between(
    "casos_menores",
    min_value=0
)

# Esta regla admite nulos por cobertura temporal del IPM
resultados["ipm_total_rango"] = validator.expect_column_values_to_be_between(
    "ipm_total",
    min_value=0,
    max_value=100,
    mostly=0.40
)

resultados["unicidad_cod_dpto_anio"] = validator.expect_compound_columns_to_be_unique(
    ["cod_dpto", "anio"]
)

# =====================================================
# 4. MOSTRAR RESULTADOS
# =====================================================
print("RESULTADOS DE VALIDACION\n")

for nombre, r in resultados.items():
    print(f"{nombre}: {r['success']}")
    if "result" in r and isinstance(r["result"], dict):
        if "unexpected_count" in r["result"]:
            print("unexpected_count:", r["result"]["unexpected_count"])
        if "element_count" in r["result"]:
            print("element_count:", r["result"]["element_count"])
    print()

# =====================================================
# 5. RESULTADO GLOBAL
# =====================================================
validaciones_criticas = [
    "cod_dpto_no_nulo",
    "dpto_no_nulo",
    "departamento_limpio_no_nulo",
    "anio_no_nulo",
    "total_casos_no_nulo",
    "anio_rango",
    "total_casos_positivo",
    "casos_femenino_positivo",
    "casos_masculino_positivo",
    "casos_adultos_positivo",
    "casos_menores_positivo",
    "unicidad_cod_dpto_anio"
]

resultado_global_critico = all(resultados[k]["success"] for k in validaciones_criticas)
resultado_ipm = resultados["ipm_total_rango"]["success"]

print("¿Pasaron todas las validaciones críticas?:", resultado_global_critico)
print("¿La validación de IPM pasó con tolerancia?:", resultado_ipm)

if resultado_global_critico:
    print("\nVALIDACION FINAL: APROBADA")
    print("El dataset puede cargarse al esquema final.")
else:
    print("\nVALIDACION FINAL: RECHAZADA")
    print("El dataset NO debe cargarse al esquema final.")
```
