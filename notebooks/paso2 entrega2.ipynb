```python
!pip install pandas matplotlib seaborn unidecode -q

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from unidecode import unidecode

# =====================================================
# 1. CARGAR ARCHIVOS GENERADOS EN EL NOTEBOOK 01
# =====================================================
df_violencia = pd.read_csv("01_violencia_raw_limpia.csv")
df_dane = pd.read_csv("02_dane_ipm_long.csv")
df_geo = pd.read_csv("03_api_divipola.csv")

print("Violencia:", df_violencia.shape)
print("DANE:", df_dane.shape)
print("Geo:", df_geo.shape)

display(df_violencia.head())
display(df_dane.head())
display(df_geo.head())

# =====================================================
# 2. FUNCION DE NORMALIZACION
# =====================================================
def normalizar_texto(x):
    if pd.isna(x):
        return None
    x = str(x).upper().strip()
    x = re.sub(r"\s*\(CT\)", "", x)
    x = re.sub(r"\s+", " ", x)
    x = unidecode(x)
    return x

# =====================================================
# 3. NORMALIZAR COLUMNAS CLAVE
# =====================================================
df_violencia["departamento_limpio"] = df_violencia["departamento"].apply(normalizar_texto)
df_violencia["municipio_limpio"] = df_violencia["municipio"].apply(normalizar_texto)

df_dane["departamento_limpio"] = df_dane["departamento"].apply(normalizar_texto)

df_geo["departamento_limpio"] = df_geo["dpto"].apply(normalizar_texto)
df_geo["municipio_limpio"] = df_geo["nom_mpio"].apply(normalizar_texto)

print("Normalización lista")
display(df_violencia[["departamento", "departamento_limpio", "municipio", "municipio_limpio"]].head())
display(df_dane[["departamento", "departamento_limpio"]].head())
display(df_geo[["dpto", "departamento_limpio", "nom_mpio", "municipio_limpio"]].head())

# =====================================================
# 4. EDA BASICO
# =====================================================
print("\n===== EDA VIOLENCIA =====")
print(df_violencia.info())
print(df_violencia.isnull().sum())
print("Duplicados:", df_violencia.duplicated().sum())

print("\n===== EDA DANE =====")
print(df_dane.info())
print(df_dane.isnull().sum())
print("Duplicados:", df_dane.duplicated().sum())

print("\n===== EDA GEO =====")
print(df_geo.info())
print(df_geo.isnull().sum())
print("Duplicados:", df_geo.duplicated().sum())

# =====================================================
# 5. LIMPIEZA DE TIPOS
# =====================================================
df_violencia["anio"] = pd.to_numeric(df_violencia["anio"], errors="coerce")
df_violencia["cantidad"] = pd.to_numeric(df_violencia["cantidad"], errors="coerce")

df_dane["anio"] = pd.to_numeric(df_dane["anio"], errors="coerce")
df_dane["ipm_total"] = pd.to_numeric(df_dane["ipm_total"], errors="coerce")

# =====================================================
# 6. AGREGAR VIOLENCIA POR DEPARTAMENTO Y AÑO
# =====================================================
violencia_agregada = (
    df_violencia
    .groupby(["departamento_limpio", "anio"], as_index=False)
    .agg(total_casos=("cantidad", "sum"))
)

print("Violencia agregada:", violencia_agregada.shape)
display(violencia_agregada.head())

# =====================================================
# 7. AGREGADOS EXTRA PARA DASHBOARD
# =====================================================
femenino = (
    df_violencia[df_violencia["genero"] == "FEMENINO"]
    .groupby(["departamento_limpio", "anio"], as_index=False)
    .agg(casos_femenino=("cantidad", "sum"))
)

masculino = (
    df_violencia[df_violencia["genero"] == "MASCULINO"]
    .groupby(["departamento_limpio", "anio"], as_index=False)
    .agg(casos_masculino=("cantidad", "sum"))
)

adultos = (
    df_violencia[df_violencia["grupo_etario"] == "ADULTOS"]
    .groupby(["departamento_limpio", "anio"], as_index=False)
    .agg(casos_adultos=("cantidad", "sum"))
)

menores = (
    df_violencia[df_violencia["grupo_etario"] == "MENORES"]
    .groupby(["departamento_limpio", "anio"], as_index=False)
    .agg(casos_menores=("cantidad", "sum"))
)

# =====================================================
# 8. TABLA DEPARTAMENTAL DESDE LA API
# =====================================================
geo_departamentos = (
    df_geo[["cod_dpto", "dpto", "departamento_limpio"]]
    .drop_duplicates()
    .reset_index(drop=True)
)

geo_departamentos["cod_dpto"] = pd.to_numeric(geo_departamentos["cod_dpto"], errors="coerce")
geo_departamentos["cod_dpto"] = geo_departamentos["cod_dpto"].astype("Int64")
geo_departamentos["cod_dpto"] = geo_departamentos["cod_dpto"].astype(str).replace("<NA>", pd.NA)
geo_departamentos["cod_dpto"] = geo_departamentos["cod_dpto"].apply(lambda x: x.zfill(2) if pd.notna(x) else x)

print("Geo departamentos:", geo_departamentos.shape)
display(geo_departamentos.head())

# =====================================================
# 9. HOMOLOGACION DE NOMBRES DEPARTAMENTALES
# =====================================================
homologacion_departamentos = {
    "GUAJIRA": "LA GUAJIRA",
    "SAN ANDRES": "SAN ANDRES, PROVIDENCIA Y SANTA CATALINA",
    "VALLE": "VALLE DEL CAUCA",
    "BOGOTA D.C.": "BOGOTA, D.C.",
    "BOGOTA DC": "BOGOTA, D.C."
}

violencia_agregada["departamento_limpio"] = violencia_agregada["departamento_limpio"].replace(homologacion_departamentos)
femenino["departamento_limpio"] = femenino["departamento_limpio"].replace(homologacion_departamentos)
masculino["departamento_limpio"] = masculino["departamento_limpio"].replace(homologacion_departamentos)
adultos["departamento_limpio"] = adultos["departamento_limpio"].replace(homologacion_departamentos)
menores["departamento_limpio"] = menores["departamento_limpio"].replace(homologacion_departamentos)
df_dane["departamento_limpio"] = df_dane["departamento_limpio"].replace(homologacion_departamentos)
geo_departamentos["departamento_limpio"] = geo_departamentos["departamento_limpio"].replace(homologacion_departamentos)

# Ajuste manual específico para San Andrés
geo_departamentos_extra = pd.DataFrame({
    "cod_dpto": ["88"],
    "dpto": ["SAN ANDRES Y PROVIDENCIA"],
    "departamento_limpio": ["SAN ANDRES, PROVIDENCIA Y SANTA CATALINA"]
})

geo_departamentos = pd.concat([geo_departamentos, geo_departamentos_extra], ignore_index=True)
geo_departamentos = geo_departamentos.drop_duplicates(subset=["departamento_limpio"], keep="first").reset_index(drop=True)

print("Homologación aplicada")
display(
    geo_departamentos[
        geo_departamentos["departamento_limpio"].str.contains("SAN ANDRES|BOGOTA", na=False)
    ]
)

# =====================================================
# 10. UNIR VIOLENCIA + DANE
# =====================================================
df_final = violencia_agregada.merge(
    df_dane[["departamento_limpio", "anio", "ipm_total"]],
    on=["departamento_limpio", "anio"],
    how="left"
)

df_final = df_final.merge(femenino, on=["departamento_limpio", "anio"], how="left")
df_final = df_final.merge(masculino, on=["departamento_limpio", "anio"], how="left")
df_final = df_final.merge(adultos, on=["departamento_limpio", "anio"], how="left")
df_final = df_final.merge(menores, on=["departamento_limpio", "anio"], how="left")

df_final = df_final.merge(
    geo_departamentos,
    on="departamento_limpio",
    how="left"
)

# =====================================================
# 11. ORDENAR COLUMNAS
# =====================================================
columnas_finales = [
    "cod_dpto",
    "dpto",
    "departamento_limpio",
    "anio",
    "total_casos",
    "casos_femenino",
    "casos_masculino",
    "casos_adultos",
    "casos_menores",
    "ipm_total"
]

df_final = df_final[columnas_finales].copy()

# =====================================================
# 12. RELLENAR NULOS EN AGREGADOS
# =====================================================
for col in ["casos_femenino", "casos_masculino", "casos_adultos", "casos_menores"]:
    df_final[col] = pd.to_numeric(df_final[col], errors="coerce").fillna(0).astype(int)

# =====================================================
# 13. AJUSTE FINAL DEFINITIVO DE cod_dpto Y dpto
# =====================================================
mapa_cod_por_nombre = (
    geo_departamentos.drop_duplicates(subset=["departamento_limpio"])
    .set_index("departamento_limpio")["cod_dpto"]
    .to_dict()
)

mapa_dpto_oficial = (
    geo_departamentos.drop_duplicates(subset=["departamento_limpio"])
    .set_index("departamento_limpio")["dpto"]
    .to_dict()
)

df_final["cod_dpto"] = df_final["cod_dpto"].fillna(
    df_final["departamento_limpio"].map(mapa_cod_por_nombre)
)

df_final["dpto"] = df_final["dpto"].fillna(
    df_final["departamento_limpio"].map(mapa_dpto_oficial)
)

df_final["cod_dpto"] = pd.to_numeric(df_final["cod_dpto"], errors="coerce")
df_final["cod_dpto"] = df_final["cod_dpto"].astype("Int64")
df_final["cod_dpto"] = df_final["cod_dpto"].astype(str).replace("<NA>", pd.NA)
df_final["cod_dpto"] = df_final["cod_dpto"].apply(lambda x: x.zfill(2) if pd.notna(x) else x)

df_final["dpto"] = df_final["dpto"].astype(str).replace("<NA>", pd.NA)
df_final["anio"] = pd.to_numeric(df_final["anio"], errors="coerce").astype("Int64")
df_final["total_casos"] = pd.to_numeric(df_final["total_casos"], errors="coerce").fillna(0).astype(int)
df_final["ipm_total"] = pd.to_numeric(df_final["ipm_total"], errors="coerce")

print("Dataset final:", df_final.shape)
display(df_final.head(10))

# =====================================================
# 14. DIAGNOSTICO FINAL
# =====================================================
print("\nNulos por columna en dataset final:")
print(df_final.isnull().sum())

print("\nFilas con cod_dpto nulo:")
display(
    df_final[df_final["cod_dpto"].isnull()][["departamento_limpio", "dpto", "anio"]]
    .drop_duplicates()
    .sort_values(["departamento_limpio", "anio"])
)

print("\nFilas con dpto nulo:")
display(
    df_final[df_final["dpto"].isnull()][["departamento_limpio", "cod_dpto", "anio"]]
    .drop_duplicates()
    .sort_values(["departamento_limpio", "anio"])
)

print("\nRegistros sin IPM:")
display(
    df_final[df_final["ipm_total"].isnull()][["departamento_limpio", "anio"]]
    .head(20)
)

# =====================================================
# 15. VISUALIZACIONES BASICAS
# =====================================================
plt.figure(figsize=(10, 5))
casos_por_anio = df_final.groupby("anio", as_index=False)["total_casos"].sum()
plt.plot(casos_por_anio["anio"], casos_por_anio["total_casos"], marker="o")
plt.title("Evolución anual de casos de violencia intrafamiliar")
plt.xlabel("Año")
plt.ylabel("Total casos")
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
top_deptos = (
    df_final.groupby("dpto", as_index=False)["total_casos"]
    .sum()
    .sort_values("total_casos", ascending=False)
    .head(10)
)
sns.barplot(data=top_deptos, x="total_casos", y="dpto")
plt.title("Top 10 departamentos con más casos")
plt.xlabel("Total casos")
plt.ylabel("Departamento")
plt.show()

df_scatter = df_final.dropna(subset=["ipm_total"]).copy()
plt.figure(figsize=(8, 6))
plt.scatter(df_scatter["ipm_total"], df_scatter["total_casos"])
plt.title("IPM vs total de casos")
plt.xlabel("IPM total")
plt.ylabel("Total casos")
plt.grid(True)
plt.show()

# =====================================================
# 16. EXPORTAR DATASET FINAL
# =====================================================
df_final.to_csv("04_dataset_final_enriquecido.csv", index=False)
print("\nArchivo guardado: 04_dataset_final_enriquecido.csv")
```
